### 1. 1부터 150까지 더한 값을 구하세요.

```python
sum_ = 0
for i in range(1,151):
    sum_ = sum_ + i
    
print(sum_)
```

### 2. 유저가 입력하는 숫자에 해당하는 단의 구구단을 보여주는 프로그램을 작성하세요.

```python
x = int(input("숫자를 입력하세요."))
print("{}단 입니다.".format(x))
for i in range(1,10):
    print("{}x{}={}".format(x,x,i,x*i))
```

### 3. 1부터 100까지 출력하되 3의 배수에는 "박수"라는 글자를 출력하는 프로그램을 작성하세요.

```python
for i in range(1,101):
    if i%3==0:
        print("박수")
    else:
        print(i)
```

### 4. 유저한테 숫자 6개를 입력 받고 그 중에 음수가 몇 개 있는지 출력하세요.


```python
list = []
for i in range(6):
    a = int(input("num"))
    if a<0:
        list.append(a)
len(list)
```

```python
cnt = 0
for i in range(6):
    num = input('숫자 입력 :')
    num = int(num)
    
    if num < 0 :
        cnt = cnt + 1
        
print('음수의 갯수는', cnt)
```

### 5. 유저한테 숫자 6개를 입력받고, 숫자들을 입력한 역순으로 출력하세요.

```python
num_list = []

for i in range(6):
    a = int(input("숫자를 입력하세요."))
    num_list.append(a)
num_list.reverse()
print(num_list)
```

### 6. 유저한테 숫자 6개를 입력받고, 숫자들을 큰 순서로 출력하세요.


```python
num_list = []

for i in range(6):
    a = int(input("숫자를 입력하세요."))
    num_list.append(a)
num_list.sort(reverse=True)
print(num_list)
```

### 7. A학습에 총 10명의 학생이 있습니다. 이 학생들의 중간고사 점수는 다음과 같습니다
### 70,60,55,75,95,90,80,80,85,76
### for문을 사용하여 A학급의 평균 점수를 구해보세요.

```python
scores = [70,60,55,75,95,90,80,80,85,76]
sum_scores = 0

for score in scores:
    sum_scores = sum_scores + score
    
avg = sum_scores/10
print(avg)
```

### 8. 유저가 점수를 입력하면 해당 학점을 출력하는 프로그램입니다.

### 단, 유저가 -1을 입력하면 프로그램을 종료합니다. -1을 입력하기 전에는 계속 입력을 받습니다.


```python
while True :
    num = int(input("점수를 입력하세요."))
    
    if num == -1:
        break
    elif 0<num<=64:
        print("F학점입니다.")
    elif 64<num<=74:
        print("C학점입니다.")
    elif 74<num<=84:
        print("B학점입니다.")
    elif 84<num<=100:
        print("A학점입니다.")
    else:
        print("잘못입력했습니다.")
```

```python
name_list = []
eng_list = []
math_list = []

for i in range(3) :
    name = input('이름 입력 : ')
    eng = input('영어 점수 :')
    math = input('수학 점수 :')
    
    name_list.append(name)
    eng_list.append(int(eng))
    math_list.append(int(math))
print(name_list)
eng_avg = sum(eng_list) / len(eng_list)
math_avg = sum(math_list) / len(math_list)
print('영어 평균은', eng_avg)
print('수학 평균은', math_avg)
```
